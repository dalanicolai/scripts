```python
WIDTH = 9
HEIGHT = 9

class Generation:
    def __init__(self, matrix):
        self.matrix = matrix

    # when this class is used!
    # def compare(self, other):
    #     # no differences
    #     result = False
    #     for x, row in enumerate(other):
    #         print 'comparing', row, self.matrix[x]
    #         # for y, cell in enumerate(row):
    #         #     if cell == self.matrix[x][y]:
    #
    #     return result

class GenerationRow:
    def __init__(self, generation):
        self.matrix = [generation]

    def add_generation(self, generation):
        self.matrix.append(generation)

    def compare(self, other):
        for x, row in enumerate(self.matrix[-1]):
            if row != other[x]:
                return False
        return True
        #
        # for i, row in enumerate(self.matrix[-1]):
        #     for j, other_row in enumerate(other):
        #         if row != other_row:
        #             return False
        # return True

    def check_periods(self):
        flipped = self.matrix[-2::-1]
        # print len(flipped), len(self.matrix)
        for x, matrix in enumerate(flipped):
            if self.compare(matrix):
                print 'exactly the same; level ', x + 1
                return x + 1
        return -1

    def evolve(self):
        new_gen = []
        for i, row in enumerate(self.matrix[-1]):
            new_gen.append([])
            for j, cell in enumerate(row):
                new_cell = cell
                living_neighbours = calculate_living_neighbours(self.matrix[-1], i, j)
                if cell == 0:
                    if living_neighbours == 3:
                        new_cell = 1
                else:
                    if living_neighbours in range(2, 4):
                        new_cell = 1
                    else:
                        new_cell = 0
                new_gen[i].append(new_cell)
        return new_gen


def draw(generation):
    for x, row in enumerate(generation.matrix[-1]):
        print ''
        for y, cell in enumerate(row):
            print cell,

def calculate_living_neighbours(generation, x, y):
    n = 0
    for cx in range(x - 1, x + 2):
        for cy in range(y - 1, y + 2):
            # all edge cases
            if cx < 0 or cy < 0 or cx >= WIDTH or cy >= HEIGHT:
                continue
            # not the same cell
            if cx == x and cy == y:
                continue
            if generation[cx][cy] == 1:
                n += 1
    return n


def parse_row_to_binary(row):
    row = row.replace(' ', '0')
    row = row.replace('x', '1')
    row = map(int, row)
    return row

def process(life_input):
    life_input = life_input.replace('\r', '').split('\n')
    max_gen = int(life_input[0])
    largest_period = int(life_input[1])
    board = life_input[2:]
    board = map(parse_row_to_binary, board)
    generations = GenerationRow(board)
    draw(generations)

    while True:
        if len(generations.matrix) <= max_gen:
            new_gen = generations.evolve()
            generations.add_generation(new_gen)
            searched_period = generations.check_periods()
            if searched_period != -1:
                print searched_period, largest_period
                if searched_period < largest_period:
                    print('Terminated. Figure has become an Oscillator. Period: %d' % searched_period)
                else:
                    print('Terminated. Figure is a Still Figure.')
                draw(generations)
        else:
            print('Terminated. Exceeded maximum number of generations; #%d' % max_gen)
            break    
```
